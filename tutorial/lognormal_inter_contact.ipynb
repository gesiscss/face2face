```python
import powerlaw
import matplotlib.pyplot as plt
import numpy as np
from sociopatterns.imports.load_all_data import Data
from sociopatterns.statistics.distribution import *
```

```python
def get_distances(delta_t_list, distance_measure="D", amount_xmin=10000):
    list_D = []
    #list_sigma = []
    #list_mu = []
    list_x = []
    for i in range(amount_xmin):
        fit = powerlaw.Fit(delta_t_list, xmin=i)
        #if np.isnan(fit.power_law.D) == False:
        list_D.append(fit.lognormal.D)
        #list_sigma.append(fit.power_law.sigma)
        #list_mu.append(fit.power_law.alpha)
        list_x.append(i)
        #else:
        #    pass
    return [list_D,list_x]
    #res = [i / j for i, j in zip(list_sigma, list_mu)] 
    
    #fig, ax1 = plt.subplots()
    #ax1.plot(list_x,list_D,linestyle='-', color='b')
    
    #ax2 = ax1.twinx()
    #ax2.plot(list_x,list_sigma,linestyle='--', color ='g')
    
    #ax3 = ax1.twinx()
    #ax3 = ax1.twiny()
    #ax3.plot(list_mu,list_sigma,linestyle='--', color='r')
    
def plot_distances(liste):
    plt.plot(liste[1],liste[0],linestyle='-', color='b')
    #plt.plot(list_x,list_sigma,linestyle='--', color ='g')
    #plt.plot(list_x,res,linestyle='--', color='r')
    plt.show()
    #return [list_D,res,list_sigma,list_mu]
```

```python
df_hospital = Data("hospital")
x_y_list_hospital, delta_t_list_hospital = calculate_inter_contact_duration(df_hospital)
liste_hospital = get_distances(delta_t_list_hospital)
```

```python
plot_distances(liste_hospital)
```
