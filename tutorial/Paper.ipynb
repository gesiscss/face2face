**Description**: For a better comprehensibility of how the figures inside the paper "The statistical characteristics of face-to-face interaction" were created here is a small tutorial. 

For the visualization we used the python library matplotlib with the settings you can see below

```python
import matplotlib.pyplot as plt
import matplotlib.ticker
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 10}

matplotlib.rc('font', **font)
plt.rc('legend', fontsize=5)
```

For the measuring and the plotting of the best fitting distributions in comparison to the original data we used the package powerlaw.

```python
import powerlaw
```

For the measurements of the different durations that were analyzed in the paper. Namely contact duration, triangle duration and inter-contact duration and for the import of the data sets that were used we imported the functions from the face2face toolbox.

```python
from face2face.statistics.distribution import *
from face2face.imports.load_all_data import Data
```

```python
liste = ["WS16","hospital","HT09","InVS13","InVS15","PS09","SFHH09","sg09","thiers11","thiers12","thiers13"]
```

For the second figure that were used in the paper the intention was to show that for all the different kinds of duration the two distributions lognormal and truncated powerlaw are pretty similar and best fitting for the datasets. You can run this code for all data sets with the loop or instead pick specific data sets.

```python
#for i in liste:
df = Data("hospital")
list_x_y_contact, delta_t_list_contact = f2f.calculate_contact_duration(df)
list_x_y_triangle, delta_t_list_triangle = f2f.calculate_triangle_duration(df)
list_x_y_inter, delta_t_list_inter = f2f.calculate_inter_contact_duration(df)
    
fit = powerlaw.Fit(delta_t_list_contact)
plt.subplot(121)
fit.plot_pdf(color='#1b9e77', linewidth=2, label="Original data", original_data=True)
fit.lognormal.plot_pdf(color='#d95f02', linestyle="--", label="Lognormal")
fit.truncated_power_law.plot_pdf(color='#7570b3', linestyle="--", label="Truncated Powerlaw")
plt.ylabel("P(${\Delta t}_{ij}$)")
plt.xlabel("${\Delta t}_{ij}$ (seconds)")
plt.legend(loc="upper right")
plt.title("contact (hospital)")
    
fit = powerlaw.Fit(delta_t_list_triangle)
plt.subplot(222)
fit.plot_pdf(color='#1b9e77', linewidth=2, label="Original data", original_data=True)
fit.lognormal.plot_pdf(color='#d95f02', linestyle="--", label="Lognormal")
fit.truncated_power_law.plot_pdf(color='#7570b3', linestyle="--", label="Truncated Powerlaw")
plt.ylabel("P(${\Delta t}_{ijk}$)")
plt.xlabel("${\Delta t}_{ijk}$ (seconds)")
plt.legend(loc="upper right")
plt.title("triangle (hospital)")
    
fit = powerlaw.Fit(delta_t_list_inter)
plt.subplot(224)
fit.plot_pdf(color='#1b9e77', linewidth=2, label="Original data", original_data=True)
fit.lognormal.plot_pdf(color='#d95f02', linestyle="--", label="Lognormal")
fit.truncated_power_law.plot_pdf(color='#7570b3', linestyle="--", label="Truncated Powerlaw")
plt.ylabel("P($t_{ij} - t_{ik}$)")
plt.xlabel("$t_{ij} - t_{ik}$ (seconds)")
plt.legend(loc="upper right")
plt.title("inter-contact (hospital)")
plt.tight_layout()
plt.show()
```
